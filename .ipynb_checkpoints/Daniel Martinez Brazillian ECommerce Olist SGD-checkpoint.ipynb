#### Como se quiere predecir la clasificaci√≥n de una producto teniendo en cuenta la ciudad, es importante tener en cuenta el id del producto, el precio, el puntaje de la rese√±a, el id del vendedor, la ciudad del usuario que hizo la rese√±a. 
##### order_status
##### order_delivered_carrier_date
##### order_delivered_customer_date
##### order_estimated_delivery_date
##### product_id
##### seller_id
##### shipping_limit_date
##### price
##### freight_value
##### review_score
##### customer_city
##### product_category_name
##### seller_city
#### Como la ciudad, el id del producto y el id del vendedor son variables categ√≥ricas, se deben convertir a variables continuas.

```python
# Se eliminan las columnas que no influyen en la calificaci√≥n de la rese√±a
olist_sellers_mg = olist_sellers_mg.drop(columns=['order_id','customer_id','order_item_id',
                                                       'review_id','review_comment_title','review_comment_message'
                                                       ,'review_creation_date','review_answer_timestamp','customer_unique_id',
                                                       'customer_zip_code_prefix','product_name_lenght'
                                                 ,'product_description_lenght','product_photos_qty','product_weight_g',
                                                 'product_length_cm','product_height_cm','product_width_cm',
                                                 'seller_zip_code_prefix'], axis=1)
```

```python
# Se calcula el npumero de d√≠as que se demora un producto en ser entregado desde su fecha de env√≠o
olist_sellers_mg['diff_estimated_approval_date'] = olist_sellers_mg.apply(lambda row:
                                                                                   np.abs(pd.to_datetime(row['order_purchase_timestamp']).day
                                                                                         - pd.to_datetime(row['order_delivered_customer_date']).day), axis=1)
```

```python
# Se eliminan las filas que tengan alg√∫n dato NaN
olist_sellers_mg = olist_sellers_mg.dropna()
```

```python
olist_sellers_mg.head(10)
```

```python
olist_sellers_mg.shape
```

```python
olist_sellers_mg.dtypes
```

```python
# Se separa el conjunto de datos para poder hacer la transformaci√≥n de los atributos categ√≥ricos a continuos para poder entrenar
# el modelo.
object_mg = olist_sellers_mg.select_dtypes(include=['object']).copy()
int_mg = olist_sellers_mg.select_dtypes(include=['int64']).copy()
float_mg = olist_sellers_mg.select_dtypes(include=['float64']).copy()

mg_int_float = pd.concat([float_mg,int_mg], axis=1)

mg_int_float.head(10)
```

```python
# Se transforman los atributos de tipo Object a num√©ricas y luego se unen con los datos num√©ricos ya existentes para tener el 
# nuevo conjunto transformado.
label_encoder = preprocessing.LabelEncoder()
object_mg_transformed = object_mg.astype(str).apply(label_encoder.fit_transform)

mg_final = pd.concat([mg_int_float,object_mg_transformed], axis=1)

mg_final.head(10)
```

```python
# Se separa el conjunto de datos
X = mg_final.drop(columns='review_score', axis=1)
y = mg_final['review_score']
```

```python
# Se preparan las variables para separar el conjunto de datos en datos de entrenamiento y datos de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
scaler = preprocessing.StandardScaler().fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
```

```python
X_train_scaled.mean(axis=0)
```

```python
X_train_scaled.std(axis=0)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled.mean(axis=0)
```

```python
X_test_scaled.std(axis=0)
```

```python
pipeline = make_pipeline(preprocessing.StandardScaler(), RandomForestRegressor(n_estimators=200, n_jobs=-1, oob_score=True))
```

```python
pipeline.get_params()
```

```python
hyperparameters = { 'randomforestregressor__max_features' : ['auto', 'sqrt', 'log2'],
                  'randomforestregressor__max_depth': [None, 5, 3, 1]}

#hyperparameters = { 'randomforestregressor__max_features' : ['auto', 'sqrt', 'log2'],
#                  'randomforestregressor__max_depth': [2,4,8,16,32,None],
#                  'randomforestregressor__n_estimators': [5,10,50,100,200,400]}
```

```python
clf = GridSearchCV(pipeline, hyperparameters, cv=10)
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf.refit
```

```python
prediction = clf.predict(X_test)
```

```python
pred=pd.DataFrame({'Actual':y_test, 'Predicted':prediction})
pred
```

```python
score = cross_val_score(clf, X_train,y_train, scoring='accuracy')
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, prediction))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, prediction))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
print('R2 Error: ', metrics.r2_score(y_test, prediction))
print("CV mean score: ", score.mean())
#print("Accuracy:",metrics.accuracy_score(y_test, prediction))

# Calculate the absolute errors
errors = abs(prediction - y_test)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')

# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / y_test)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)

print('Accuracy:', round(accuracy, 2), '%.')
#get the mean of each fold 
print("Accuracy of Model with Cross Validation is:",score.mean() * 100)
```
